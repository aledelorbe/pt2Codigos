### Contar la frecuencia de cada valor distinto en la columna ICD

```python
dfDefinitivo = pd.concat([df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019], ignore_index=True)

valores_distintos = dfDefinitivo['ICD'].value_counts(dropna=False)
nuevo_dataframe = pd.DataFrame({'llave': valores_distintos.index, 'Frecuencia': valores_distintos.values})
print(nuevo_dataframe)
```

### Armar la "tabla" que vendr√≠a a ser la descripci√≥n de las llaves c√°ncer

```python
# Se toma el conjunto de 2011 para que salgan los 36 valores distintos y el valor extra (valor nulo) que sale en este dataset

df2011.drop_duplicates(subset='ICD')[['ICD', 'organ']].value_counts(dropna=False).count() # Los 36
descripCancer = df2011.drop_duplicates(subset='ICD')[['ICD', 'organ']]
print(descripCancer)
```

### Ahora si hacer el join con el dataset que tiene la descripci√≥n del tipo de c√°ncer

```python
dfJoin = pd.merge(nuevo_dataframe, descripCancer, left_on='llave', right_on='ICD', how='inner')
print(dfJoin)
```

### Graficar resultado

```python
plt.figure(figsize=(12, 6))
plt.bar(dfJoin.index, dfJoin['Frecuencia'], color='#17becf', width=0.7)
plt.title('Cantidad de personas en cada categor√≠a del c√°ncer en M√©xico en los a√±os 2010 - 2019')
plt.xlabel('Nombre de la categor√≠a del c√°ncer')
plt.ylabel('Frecuencia')

plt.xticks(dfJoin.index, dfJoin['organ'], rotation=90)
plt.grid(True)
```
